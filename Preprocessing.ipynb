```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('PreLibBooks.csv')
```

```python
df.head(1200)
```

```python
df['S. NO.'] = df.reset_index().index + 1
```

```python
# Count duplicates and update numCopies column
df['numCopies'] = df.groupby('NAME OF BOOK')['NAME OF BOOK'].transform('count')
```

```python
df.sort_values('numCopies', ascending = False)
```
